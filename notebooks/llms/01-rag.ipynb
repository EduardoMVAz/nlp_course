## Exercise 2

```python
from collections import defaultdict

def find_cells_with_keys(keys: list, notebook: dict):
    cells = defaultdict(list)

    for i, cell in enumerate(notebook["cells"]):
        for key in keys:
            if key in " ".join(cell["source"]).lower():
                cells[i].append(key)

    return dict(cells)
```

```python
keys = [
    "llama",
    "artificial",
    "words",
    "learning",
    "python",
    "regular",
    "expression"
]
```

```python
find_cells_with_keys(keys, dict(notebook_data))
```

## Exercise 3

```python
from transformers import BertTokenizer, BertModel

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained("bert-base-uncased")
```

```python
def get_embeddings(text, model, tokenizer):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    outputs = model(**inputs)
    cls_embedding = outputs.last_hidden_state[0, 0, :]
    return cls_embedding

def get_notebook_embeddings(notebook: dict, model, tokenizer):
    cell_embeddings = defaultdict()

    for i, cell in enumerate(notebook["cells"]):
        cell_embedding = get_embeddings(" ".join(cell["source"]), model, tokenizer)
        cell_embeddings[i] = cell_embedding

    return cell_embeddings
```

```python
phrase = "What is a regular expression?"
```

```python
user_phrase_embeddings = get_embeddings(phrase, model, tokenizer)
```

```python
notebook_embeddings = get_notebook_embeddings(dict(notebook_data), model, tokenizer)
```
